$R = \sqrt{x^2 + y^2}$

But here we slightly modified it:

$R^2 = (x - x_{center})^2 + (y - y_{center})^2$

```python
w = h = 11
grid = ImageGrid(w, h, block_size=50)
grid

for pixel in grid:
    if ((pixel.x - 5) ** 2 + (pixel.y - 5) ** 2) < 5 ** 2:
        pixel.green = 255

grid
```

```python
from mobilechelonian import Turtle
```

Following methods of turtle object are available:
* `forward(X)` moves turtle to X steps in current direction
* `left(Y)` and `right(Y)` turns turtle left or right to Y degrees (from 0 to 360)

Here example code that draws $\infty$ symbol

```python
t = Turtle()
t.speed(10)

t.left(45)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(200)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
```
