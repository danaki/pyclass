# Booleans

Represented by two constants: **True** or **False**

```python
True, False
```

the type is called `bool`

```python
type(True)
```

### Comparison operators produce Booleans

* `==`
* `is`
* `>`, `<`, `>=`, `<=`

Produce Boolean by comparing two arguments

```python
1 < 2
```

#### Logical operators (not, and, or)

`not` inverts the Boolean value

```python
not True
```

`and` produces True **only** if both operands are True, False otherwise



`or` produces True if at least one argument is True, False otherwise (actually it's only when both are False)



You can build formulas using Boolean logic like so:

```python
print True and True and True and False
print False or False or False or True
```

but remember about...

### Operator precedence

* not
* and
* or

so in an expression:

```python
True or False and not True
```

* `not False` is going to be evaluated first
* then `False and "not False"` will be evaluated next
* and `True or "False and not False"` will be computed last

You can always change the order of operations by enclosing parts using parenthesis, like so:

```python
(True or False) and not True
```
