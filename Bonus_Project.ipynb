````python
import matplotlib.lines as lines
import matplotlib.pyplot as plt
import math
%matplotlib inline
````

Look at the function `plot()` in the cell below, it accepts two arguments:
 1. `vertices` is the list of circle points. Each elements of this list is a tuple of point coordinates: (x, y)
 1. `edges` is the list of vertix indices to be connected, each element is a tuple. For example: value (7, 8) connects vertices #7 and #8
 
You will call this function to plot your own solution later. No need to modify this function.

````python
def plot(vertices, edges, details=False, corner=None):
    fig, ax = plt.subplots()
    fig.set_size_inches(10, 10)
    plt.scatter([-1, 1], [-1, 1], s=0)
    
    if details:
        # draw vertix indices        
        for i, (vx, vy) in enumerate(vertices):
            plt.text(vx * 1.05, vy * 1.05, i, fontsize=12, horizontalalignment='center', verticalalignment='center')
    else:
        plt.axis('off')
    
    if corner:
        plt.text(-1, -1, corner, fontsize=12)        
    
    # draw edges
    for edge in edges:
        # Note that the Line2D takes a list of x values and a list of y values,
        # not 2 points as one might expect. So we have to convert our points
        # an x-list and a y-list.
        (line_xs, line_ys) = zip(vertices[edge[0]], vertices[edge[1]])
        ax.add_line(lines.Line2D(line_xs, line_ys, linewidth=2, color='blue'))
        
    plt.plot()
    plt.show()
````

````python
from matplotlib.patches import Wedge
from matplotlib.collections import PatchCollection
from ipywidgets import interact, FloatSlider

def pacman(theta):
    fig, ax = plt.subplots()    
    angle = theta / math.pi * 180
    wedge = Wedge((0, 0), 1, 0, angle)
    ax.add_collection(PatchCollection([wedge]))

    plt.axis('equal')
    plt.axis('off')
    plt.tight_layout()

    plt.show()

interact(pacman, theta=FloatSlider(min=0, max=2 * math.pi, step=math.pi/6));
````
