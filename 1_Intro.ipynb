````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 2 * np.pi, 0.1)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
plt.plot(x, y)
plt.show()
````

## Variables

Variables can be defined like this:

````python
# one-letter variable names are often used in local short-living context
a = 1.0123

# i, j, k... are good name for iterators
i = 1

# you can use alphabet letters, digits or underscore symbol
# however an identifier cannot start with a digit, so `7days` is an invalid name
my_message = "Hi there!"

# bad: variable name is excessively descriptive
students_with_grades_above_90 = ['John', 'Jack', 'Gerom']
````

Treat variables as tags for values. Both `var1` and `var2` are referrencing the same object. This is usually not a problem because strings are immutable.



This behavior can make your code more performant and look more or less readable, however it can also lead to bugs.
