## Exercise: Sum of Multiples

Given a number, find the sum of all the unique multiples of particular numbers up to but not including that number.

If we list all the natural numbers below 20 that are multiples of 3 or 5, we get 3, 5, 6, 9, 10, 12, 15, and 18.

The sum of these multiples is 78.

````python
def sum_of_multiples(n, mutliples):
    accumulator = []
    for i in range(1, n):
        for m in mutliples:
            if m > 0 and i % m == 0:
                accumulator.append(i)
                break
    
    #print accumulator
    return sum(accumulator)

assert sum_of_multiples(1, [3, 5]) == 0
assert sum_of_multiples(4, [3, 5]) == 3
assert sum_of_multiples(7, [3]) == 9
assert sum_of_multiples(10, [3, 5]) == 23
assert sum_of_multiples(100, [3, 5]) == 2318
assert sum_of_multiples(1000, [3, 5]) == 233168
assert sum_of_multiples(20, [7, 13, 17]) == 51
assert sum_of_multiples(15, [4, 6]) == 30
assert sum_of_multiples(150, [5, 6, 8]) == 4419
assert sum_of_multiples(51, [5, 25]) == 275
assert sum_of_multiples(10000, [43, 47]) == 2203160
assert sum_of_multiples(100, [1]) == 4950
assert sum_of_multiples(10000, []) == 0
assert sum_of_multiples(1, [0]) == 0
assert sum_of_multiples(4, [0, 3]) == 3
assert sum_of_multiples(10000, [2, 3, 5, 7, 11]) == 39614537
````

## Exercise: Transpose matrix

Given list of lists (square matrix):

```
matrix = [[1, 2, 3],
          [0, 0, 3],
          [9, 0, 3]]
```

transpose it:

$F_{i,j} = F_{j,i}$

e.g. change places columns and rows like so:

```
transp = [[1, 0, 9],
          [2, 0, 0],
          [3, 3, 3]]
```


````python
def transpose(matrix):
    rows = len(matrix)
    columns = len(matrix[0])
    
    result = [[0] * rows for i in range(columns)]
    for i in range(rows):
        for j in range(columns):
            result[j][i] = matrix[i][j]
        
    return result

assert transpose([[1, 2, 3],
                  [0, 0, 3],
                  [9, 0, 3]]) \
                     == \
                 [[1, 0, 9],
                  [2, 0, 0],
                  [3, 3, 3]]

assert transpose([[1, 2, 3, 0],
                  [0, 0, 3, 0],
                  [9, 0, 3, 0]]) \
                     == \
                 [[1, 0, 9],
                  [2, 0, 0],
                  [3, 3, 3],
                  [0, 0, 0]]
````
